#### reward changes in MediaLogo condition

```python
rewardssignature_path = '/Volumes/Shaolin/toolboxes_menghan/material/masks/rewardssignature'
rewardssignature = Brain_Data(os.path.join(rewardssignature_path,'SVM_HCP_Reward_vs_Punish_n392_zscore1_smooth0_weightmap.nii.gz'))
rewardssignature.plot()

rewardssignature_masked = Brain_Data(os.path.join(rewardssignature_path,'SVM_HCP_Reward_vs_Punish_n392_masked_zscore1_smooth0_weightmap.nii.gz'))
rewardssignature_masked.plot()
```

```python


i = condition[2]
rewards_ROIs = []


rewards_ROIs_all = pd.DataFrame([])
rewards_ROIs_masked_all = pd.DataFrame([])
for j in tqdm(run):
    rewards_ROIs = pd.DataFrame([])
    rewards_ROIs_masked = pd.DataFrame([])
    group_file_list = glob.glob(os.path.join(base_dir,'denoised','smoothed_runs','beta',f'sub*{i}*run-{j}.nii.gz'))
    group_file_list.sort()
    group_dat_ROI = Brain_Data(group_file_list)

    corr = group_dat_ROI.similarity(rewardssignature)
    corr_masked = group_dat_ROI.similarity(rewardssignature_masked)
    
    rewards_ROIs['rewards'] = corr
    rewards_ROIs['runs'] = int(j)
    rewards_ROIs_all = pd.concat([rewards_ROIs_all,rewards_ROIs],axis = 0)

    rewards_ROIs_masked['rewards'] = corr_masked
    rewards_ROIs_masked['runs'] = int(j)
    rewards_ROIs_masked_all = pd.concat([rewards_ROIs_masked_all,rewards_ROIs_masked],axis = 0)

```

```python
sns.barplot(x = 'runs',y = 'rewards',data = rewards_ROIs_all)
```
